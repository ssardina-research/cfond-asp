```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotnine import *
import plotnine
```

```python
# set the path to the csv file
inter_csv_file = "./data/controllers/interim.csv"
final_csv_file = "./data/controllers/final.csv"
configs_csv_file = "./data/controllers/configs.csv"
df_inter = pd.read_csv(inter_csv_file)
df_final = pd.read_csv(final_csv_file)
df_configs = pd.read_csv(configs_csv_file)
df_configs["asp"] = 'asp-' + df_configs['num'].astype(str)

df_inter = df_inter.join(df_configs.set_index('asp'), on='configuration')
df_final= df_final.join(df_configs.set_index('asp'), on='configuration')
pd.unique(df_inter["domain"])
```

```python
def plot_solved_iterations(scenario, metric):
    df_1 = df_inter.query(f"domain==@scenario")
    plotnine.options.figure_size = (10, 4)
    p = ggplot(df_1, aes(x="iteration", y=metric, color="factor(model)")) + geom_line() + facet_wrap('instance', scales="free") + labs(title="Solved instances") + theme(subplots_adjust={'wspace': 0.1})
    return p
```

### Beam Walk

```python
scenario = "beam-walk"
metric = "time"
p1 = plot_solved_iterations(scenario, metric)

display(p1)
```
