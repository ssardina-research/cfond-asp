```python
# import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotnine import *
import plotnine
```

```python
# set the path to the csv file
inter_csv_file = "./data/interim.csv"
timedout_csv_file = "./data/timeout.csv"
final_csv_file = "./data/final.csv"
df_solved = pd.read_csv(inter_csv_file)
df_timedout = pd.read_csv(timedout_csv_file)
df_final = pd.read_csv(final_csv_file)
```

```python
solvers = ["asp-opt-1", "asp-plus-opt-1", "fondsat-glucose", "fondsat-minisat"]
def plot_solved_iterations(scenario, solvers, metric):
    df_1 = df_solved.query(f"scenario==@scenario and solver in @solvers")
    plotnine.options.figure_size = (10, 4)
    p = ggplot(df_1, aes(x="iteration", y=metric, color="factor(solver)")) + geom_line() + facet_wrap('instance', scales="free") + labs(title="Solved instances")  + theme(subplots_adjust={'wspace': 0.1}) + theme(subplots_adjust={'hspace': 0.1})
    return p

def plot_unsolved_iterations(scenario, solvers, metric):
    df_1 = df_timedout.query(f"scenario==@scenario and solver in @solvers")
    plotnine.options.figure_size = (10, 6)
    p = ggplot(df_1, aes(x="iteration", y=metric, color="factor(solver)")) + geom_line() + facet_wrap('instance', scales="free") + labs(title="Unsolved instances") + theme(subplots_adjust={'wspace': 0.1})+ theme(subplots_adjust={'hspace': 0.1})
    return p
```

```python
scenario = "blocksworld-ipc08"
instance = "p07"
metric = "solve-time"
df_1 = df_solved.query(f"scenario==@scenario and solver in @solvers and instance==@instance")
plotnine.options.figure_size = (10, 4)
p = ggplot(df_1, aes(x="iteration", y=metric, color="factor(solver)")) + geom_line() 
p
```
