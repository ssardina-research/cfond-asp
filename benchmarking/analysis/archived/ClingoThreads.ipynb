```python
import pandas as pd
from plotnine import *
import plotnine
import numpy as np
%matplotlib inline
```

## Instance properties

`instance_states = {1: [8, 9], 2: [30, 31], 3: [9, 10], 4: [13, 14]}`

```python
df = pd.read_csv("clingo_threads.csv")

## Add column to differentiate sat and unsat instances
cond_1 = ((df.instance==1) & (df.num_states==8)) 
cond_2 = ((df.instance==2) & (df.num_states==30))
cond_3 = ((df.instance==3) & (df.num_states==9))
cond_4 = ((df.instance==4) & (df.num_states==13))
df["solution"] = np.where((cond_1 | cond_2 | cond_3 | cond_4), "unsat", "sat")
df_sat = df.query("solution=='sat'")
df_unsat = df.query("solution=='unsat'")
```

```python
plotnine.options.figure_size = (10, 6)
ggplot(df_unsat) + geom_boxplot(aes(x='factor(num_threads)', y='solve_time')) + labs(title="Unsatisfiable Instance") + facet_wrap('instance', scales = "free") + ylab("Solve time") + xlab("#Threads") + theme(subplots_adjust={'wspace': 0.1, 'hspace': 0.1}) 
```

```python
plotnine.options.figure_size = (10, 6)
ggplot(df_sat) + geom_boxplot(aes(x='factor(num_threads)', y='solve_time')) + labs(title="Satisfiable Instance") + facet_wrap('instance', scales = "free") + ylab("Solve time") + xlab("#Threads") + theme(subplots_adjust={'wspace': 0.1, 'hspace': 0.1}) 
```
